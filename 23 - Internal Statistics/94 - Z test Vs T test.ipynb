# Z-Test vs T-Test

Choosing between a Z-test and a T-test is a common decision in hypothesis testing. The choice depends on the sample size and whether the population standard deviation is known.

## Comparison Table

| Feature | Z-Test | T-Test |
| :--- | :--- | :--- |
| **Population Variance ($\sigma^2$)** | Known | Unknown |
| **Sample Size ($n$)** | Large ($n \ge 30$) | Small ($n < 30$) |
| **Distribution** | Standard Normal (Z) | Student's t |
| **Degrees of Freedom** | Not applicable | Applicable ($df = n-1$) |

## Decision Flowchart
1.  **Is Population Standard Deviation ($\sigma$) known?**
    *   **Yes:** Use **Z-Test**.
    *   **No:** Go to step 2.
2.  **Is Sample Size large ($n \ge 30$)?**
    *   **Yes:** You *can* use Z-test (because t-distribution approximates Z), but T-test is still technically more correct. In practice, Z-test is often used for simplicity.
    *   **No:** Use **T-Test**.

## Summary
*   **Z-test:** Ideal world (we know population parameters).
*   **T-test:** Real world (we estimate parameters from sample).
