# Types of ANOVA

## 1. One-Way ANOVA
*   **Definition:** Compares the means of three or more independent groups based on **one independent variable** (factor).
*   **Example:** Comparing test scores of students from 3 different schools (Factor: School).

## 2. Two-Way ANOVA
*   **Definition:** Compares the means of groups based on **two independent variables** (factors). It also checks for the **interaction effect** between the two factors.
*   **Example:** Comparing test scores based on School (Factor 1) AND Gender (Factor 2).
    *   Does School affect scores?
    *   Does Gender affect scores?
    *   Does the effect of School depend on Gender? (Interaction)

## 3. N-Way ANOVA
*   Generalization to more than two factors.

## 4. Repeated Measures ANOVA
*   Used when the same subjects are used for each treatment (like a paired t-test but for >2 treatments).
*   **Example:** Measuring blood pressure of patients at Time 1, Time 2, and Time 3.
