# Assumptions of ANOVA

For the results of an ANOVA test to be valid, certain assumptions must be met.

## 1. Normality
*   The dependent variable should be approximately normally distributed for each category of the independent variable.
*   **Check:** Shapiro-Wilk test or Q-Q plots.

## 2. Homogeneity of Variance (Homoscedasticity)
*   The variances of the different groups should be equal.
*   **Check:** Levene's test or Bartlett's test.

## 3. Independence
*   The observations in each group are independent of each other and the observations in other groups.
*   **Check:** Study design (Random sampling).

## What if assumptions are violated?
*   **Normality:** ANOVA is robust to moderate violations if sample sizes are equal. If severe, use Kruskal-Wallis H Test (Non-parametric).
*   **Homogeneity:** Use Welch's ANOVA.
