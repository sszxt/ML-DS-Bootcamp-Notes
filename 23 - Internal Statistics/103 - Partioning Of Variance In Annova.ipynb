# Partitioning of Variance in ANOVA

ANOVA works by splitting the total variation in the data into different components.

## Total Sum of Squares (SST)
The total variation in the data.
$$SST = \sum (x_{ij} - \bar{x}_{grand})^2$$

## Sum of Squares Between Groups (SSB)
The variation due to the interaction between the groups (The "Signal").
$$SSB = \sum n_j (\bar{x}_j - \bar{x}_{grand})^2$$
Where $n_j$ is the size of group $j$ and $\bar{x}_j$ is the mean of group $j$.

## Sum of Squares Within Groups (SSW) or Error (SSE)
The variation within each group (The "Noise").
$$SSW = \sum (x_{ij} - \bar{x}_j)^2$$

## The Relationship
$$SST = SSB + SSW$$

## F-Statistic Calculation
1.  **Mean Square Between (MSB):** $SSB / (k - 1)$
2.  **Mean Square Within (MSW):** $SSW / (N - k)$
3.  **F-Ratio:** $F = \frac{MSB}{MSW}$

If MSB is significantly larger than MSW, it means the group differences are larger than random noise, leading to a significant F-test.
