# Type I and Type II Errors

In hypothesis testing, we can make two types of errors when deciding whether to reject the null hypothesis ($H_0$).

## Confusion Matrix of Hypothesis Testing

| | **$H_0$ is Actually True** | **$H_0$ is Actually False** |
| :--- | :--- | :--- |
| **Reject $H_0$** | **Type I Error** ($\alpha$) <br> False Positive | **Correct Decision** ($1-\beta$) <br> Power of Test |
| **Fail to Reject $H_0$** | **Correct Decision** ($1-\alpha$) <br> Confidence Level | **Type II Error** ($\beta$) <br> False Negative |

## Type I Error ($\alpha$)
*   **Definition:** Rejecting the null hypothesis when it is actually true.
*   **AKA:** False Positive.
*   **Probability:** The significance level $\alpha$ (e.g., 0.05).
*   **Example:** Convicting an innocent person.

## Type II Error ($\beta$)
*   **Definition:** Failing to reject the null hypothesis when it is actually false.
*   **AKA:** False Negative.
*   **Probability:** $\beta$.
*   **Example:** Letting a guilty person go free.

## Trade-off
*   Reducing $\alpha$ (e.g., from 0.05 to 0.01) makes it harder to reject $H_0$, which reduces Type I errors but increases the risk of Type II errors ($\beta$).
*   Increasing sample size can reduce both errors.
