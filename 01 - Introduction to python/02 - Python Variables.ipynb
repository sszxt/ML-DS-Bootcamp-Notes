<b >PYTHON VARIABLES : </b>


Python variables have no commands for declaring the variables. The variables can be created the moment we assign a variable to it. We do not need to declare datatypes to variables. 

```python
name = "sameer"; age = 10; 
print(name) 
print(age)
```

<b >CASTING : </b>

 Casting means assigning a specific datatypes to that variable.

```python
name = str("sameer")
age = int(10)
print(name)
print(age)
```
