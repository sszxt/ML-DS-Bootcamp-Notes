<b> Python Operators : </b>


Operators are used to perform operations on variables and values. Python divides the operators into following : 
- Arithmetic Operators<br>
- Assignment Operators<br>
- Comparision Operators<br>
- Logical Operators<br>
- Identity Operators<br>
- Membership Operators<br>
- Bitwise Operators<br>

<b> Arithmetic Operators : </b> 

Arithmetic Operators are used with numeric values to perform mathematical operations like addition, subraction e.t.c.

```python
# --------------------------------------------
# Arithmetic Operators in Python (with examples)
# --------------------------------------------

a = 15
b = 4

# 1. Addition (+)
result_add = a + b          # 15 + 4 = 19
print("Addition:", result_add)

# 2. Subtraction (-)
result_sub = a - b          # 15 - 4 = 11
print("Subtraction:", result_sub)

# 3. Multiplication (*)
result_mul = a * b          # 15 * 4 = 60
print("Multiplication:", result_mul)

# 4. Division (/)
result_div = a / b          # 15 / 4 = 3.75 (always float)
print("Division:", result_div)

# 5. Floor Division (//)
result_floordiv = a // b    # 15 // 4 = 3 (drops decimal)
print("Floor Division:", result_floordiv)

# 6. Modulus (%)
result_mod = a % b          # 15 % 4 = 3 (remainder)
print("Modulus:", result_mod)

# 7. Exponent (**)
result_exp = a ** b         # 15 ** 4 = 50625 (15^4)
print("Exponent:", result_exp)
```

<b> Assignment Operators </b>

Assignment operators are used to assign values to the variables.

```python
# --------------------------------------------
# Assignment Operators in Python (with examples)
# --------------------------------------------

x = 10   # Simple assignment
print("Initial x:", x)

# 1. Add and assign (+=)
x += 5    # x = x + 5  → 10 + 5 = 15
print("After x += 5:", x)

# 2. Subtract and assign (-=)
x -= 3    # x = x - 3  → 15 - 3 = 12
print("After x -= 3:", x)

# 3. Multiply and assign (*=)
x *= 2    # x = x * 2 → 12 * 2 = 24
print("After x *= 2:", x)

# 4. Divide and assign (/=)
x /= 4    # x = x / 4 → 24 / 4 = 6.0 (becomes float)
print("After x /= 4:", x)

# 5. Floor division and assign (//=)
x //= 2   # x = x // 2 → 6.0 // 2 = 3.0
print("After x //= 2:", x)

# 6. Modulus and assign (%=)
x %= 2    # x = x % 2 → remainder of 3.0 % 2 = 1.0
print("After x %= 2:", x)

# Reset x for next examples
x = 3
print("\nReset x to:", x)

# 7. Exponent and assign (**=)
x **= 3   # x = x ** 3 → 3^3 = 27
print("After x **= 3:", x)

# 8. Bitwise AND and assign (&=)
x &= 5    # x = x & 5 (bitwise AND)
print("After x &= 5:", x)

# 9. Bitwise OR and assign (|=)
x |= 2    # x = x | 2 (bitwise OR)
print("After x |= 2:", x)

# 10. Bitwise XOR and assign (^=)
x ^= 1    # x = x ^ 1 (bitwise XOR)
print("After x ^= 1:", x)

# 11. Bitwise right shift and assign (>>=)
x >>= 1   # x = x >> 1 (right shift bits)
print("After x >>= 1:", x)

# 12. Bitwise left shift and assign (<<=)
x <<= 2   # x = x << 2 (left shift bits)
print("After x <<= 2:", x)
```

<b> Comparison Operators </b>


Comparison Operators are used to compare two values with each other.

```python
# --------------------------------------------
# Comparison Operators in Python (with examples)
# --------------------------------------------

a = 10
b = 20

print("a =", a, "| b =", b, "\n")

# 1. Equal to (==)
print("a == b:", a == b)      # Checks if a is equal to b

# 2. Not equal to (!=)
print("a != b:", a != b)      # Checks if a is not equal to b

# 3. Greater than (>)
print("a > b:", a > b)        # True if a is greater than b

# 4. Less than (<)
print("a < b:", a < b)        # True if a is less than b

# 5. Greater than or equal to (>=)
print("a >= b:", a >= b)      # True if a is greater than or equal to b

# 6. Less than or equal to (<=)
print("a <= b:", a <= b)      # True if a is less than or equal to b)

# --------------------------------------------
# Extra Comparison for same-value variables
# --------------------------------------------

x = 10
y = 10

print("\nx =", x, "| y =", y, "\n")

print("x == y:", x == y)      # True (values same)
print("x >= y:", x >= y)      # True (equal)
print("x <= y:", x <= y)      # True (equal)
```

<b> Logical operators </b>

<b>and</b> : Returns true if both statement are true. <br>
<b>or</b> : returns true if one of the statement is true. <br>
<b>not</b> : Reverse the result, returns true if the result is false and vice versa.<br>

```python
# --------------------------------------------
# Logical Operators in Python
# --------------------------------------------

a = True
b = False

# AND
print("a and b:", a and b)   # True only if both are True

# OR
print("a or b:", a or b)     # True if at least one is True

# NOT
print("not a:", not a)       # Inverts True → False
```

<b>Identity Operators : </b>

<b>is</b> : Returns true if both variables are same object. <br>
<b>is not</b> : Returns true if both variables are not same object.

```python
# --------------------------------------------
# Identity Operators in Python
# --------------------------------------------

x = [1, 2, 3]
y = [1, 2, 3]
z = x  # z references the same object as x

print("x is y:", x is y)           # False → different objects in memory
print("x == y:", x == y)           # True → values are same
print("x is z:", x is z)           # True → same memory reference

print("x is not y:", x is not y)   # True
```

<b>Membership Operators : </b>

<b>in</b> : Returns true if sequence with specified value is present in the object. <br>
<b>not in</b> : Returns true if sequence with specified value is not present in the object.

```python
# --------------------------------------------
# Membership Operators in Python
# --------------------------------------------

numbers = [10, 20, 30, 40]

print("20 in numbers:", 20 in numbers)        # True
print("50 not in numbers:", 50 not in numbers) # True

text = "hello world"
print("'hello' in text:", "hello" in text)     # True
```

<b>Bitwise Operators</b>

<b>AND</b> <b>OR</b> <b>XOR</b> <b>NOT</b> 


```python
# --------------------------------------------
# Bitwise Operators in Python
# --------------------------------------------

a = 5   # 0101
b = 3   # 0011

print("a & b:", a & b)    # AND  -> 0001 -> 1
print("a | b:", a | b)    # OR   -> 0111 -> 7
print("a ^ b:", a ^ b)    # XOR  -> 0110 -> 6
print("~a:", ~a)          # NOT  -> -(a+1) = -6
print("a << 1:", a << 1)  # Left shift  -> 1010 -> 10
print("a >> 1:", a >> 1)  # Right shift -> 0010 -> 2
```
