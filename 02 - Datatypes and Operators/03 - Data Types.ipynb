<b> DATA TYPES : </b>


In programming, Datatypes are importand concept. The variables can be store different types of data in it. The Programmers may/may not assign data types to the variables depends on the requirement of the program. But the python can identify the datatype is nothing is declared (casted) in it.

Text type : str <br>
Numeric Type ; int, float, complex <br>
Sequence type : list, tuple, range <br>
Mapping type : dict <br>
Set types : set, frozenset <br>
Binary Types : bytes, bytearray, memoryview <br>
None type : Nonetype<br>

To get the datatype of any object, we can use the type() function,

```python
number = 5
print(type(number))
```
