<b>What is module?</b>

Module is a file containing python code which can include definitions, statements, functions. It is a way to reuse a segment of code (or) function again and again.

For Example, let's take random module.

<b>Random Module</b>

The random module is used to generate random numbers and random stuffs. To use the random module, we have to import the random module to the codebase. Python's random module has lot of functions like randint() and so on, we can use whatever function we want based on our use cases. 

```python
import random

random_int = random.randint(1, 100)
print(random_int)
```
