<b>Python Dictonaries</b>

Dictonaries are used to store values in the <b>key : value</b> pairs. It is a collection which is ordered, changeable and do not allow duplicates.

```python
car = {
    "brand" : "Ford",
    "model" : "Mustang"
}
print(car)
```

<b> Dictionary Items: </b>


Dictionary items are ordered, changeable, and do not allow duplicates.
Dictionary items are presented in key:value pairs, and can be referred to by using the key name.

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(thisdict["brand"])
```
